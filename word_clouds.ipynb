```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('IMDB.csv')
pre_processed = data.head(100)
```

```python
all_rev = ' '.join(pre_processed['review'])
```

```python
stopwords = set(STOPWORDS)
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(all_rev)
```

```python
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```

```python
# only positive reviews pre processing 
```

```python
all_positive = pre_processed[pre_processed["sentiment"] == "positive"]
all_positive_text = ' '.join(all_positive['review'])    
```

```python
stopwords = set(STOPWORDS)
wordcloud_positive = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(all_positive_text)

plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud_positive)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```

```python
all_negative = pre_processed[pre_processed["sentiment"] == "negative"]
all_negative_text= ' '.join(all_negative['review'])   
stopwords = set(STOPWORDS)
wordcloud_negative = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(all_negative_text)

plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud_negative)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
