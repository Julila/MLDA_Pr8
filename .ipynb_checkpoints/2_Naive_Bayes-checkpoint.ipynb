```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
from time import time
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
# Loading pre-processed data / Creating training and test data sets
```

```python
training_data = pd.read_csv('cleaned_training_data.csv') # load cleaned data
```

```python
training_data
```

```python
test_csv = training_data.tail(10000)
train_csv = training_data.head(40000)

train_X = train_csv['lemmatization']    
train_y = train_csv["sentiment"]   
test_X = test_csv['lemmatization']
test_y = test_csv['sentiment']
```

```python
# loading CountVectorizer
tf_vectorizer = TfidfVectorizer(stop_words='english') # create Object of class TfidfVectorizer


X_train_tf = tf_vectorizer.fit_transform(train_X) # run method fit_transform on training set

  
print("n_samples: %d, n_features: %d" % X_train_tf.shape)
 
```

```python
for index, value in train_y.items():

    print(f"Index : {index}, Value : {value}")
```

```python
X_train_tf
```

```python
naive_bayes_classifier = MultinomialNB() # create object naive_bayes_classifier
naive_bayes_classifier.fit(X_train_tf, train_y) # run method fit from the class MultinomialNB


# TODO for just Bayes Classifier 
 
```

```python
# predict the new document from the testing dataset
X_test_tf = tf_vectorizer.transform(test_X) # Create a Vector TF from test data 
y_pred = naive_bayes_classifier.predict(X_test_tf)
 
# compute the performance measures
score1 = metrics.accuracy_score(test_y, y_pred)
print("accuracy:   %0.3f" % score1)

print(metrics.classification_report(test_y, y_pred,
                                            target_names=['Positive', 'Negative']))

print("confusion matrix:")
print(metrics.confusion_matrix(test_y, y_pred))

print('------------------------------')
```

```python
from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train_tf, train_y)
 
 
```

```python
# predict the new document from the testing dataset
 
y_pred = neigh.predict(X_test_tf)
 
# compute the performance measures
score1 = metrics.accuracy_score(test_y, y_pred)
print("accuracy:   %0.3f" % score1)

print(metrics.classification_report(test_y, y_pred,
                                            target_names=['Positive', 'Negative']))

print("confusion matrix:")
print(metrics.confusion_matrix(test_y, y_pred))

print('------------------------------')
```
