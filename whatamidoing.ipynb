```python
data = pd.read_csv('cleaned_data.csv') # load cleaned data
data["sentiment_int"]=data["sentiment"].replace("positive", 1).replace("negative",0)
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data['lemmatization'], data["sentiment_int"], train_size = 0.75)
```

```python
tf_vectorizer = TfidfVectorizer(stop_words='english', min_df=5, max_df=0.3) # create Object of class TfidfVectorizer

X_train_tf = tf_vectorizer.fit_transform(X_train) # run method fit_transform on training set

print("n_samples: %d, n_features: %d" % X_train_tf.shape)
 
```

```python
#X_train_tr=X_train_tf.transpose()
X_train_df = pd.DataFrame(X_train_tf.toarray(),columns=tf_vectorizer.get_feature_names())
X_train_df
```

```python
#pip install pyspellchecker
```

```python
X_test_tf = tf_vectorizer.fit_transform(X_test) # run method fit_transform on training set
X_test_df = pd.DataFrame(X_test_tf.toarray(),columns=tf_vectorizer.get_feature_names())
X_test_df
```
